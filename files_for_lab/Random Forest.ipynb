```python
target.TARGET_B.value_counts(dropna=False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train_df = pd.DataFrame(X_train, columns=X.columns)
X_test_df  = pd.DataFrame(X_test,  columns=X.columns)

X_train_num = X_train_df.select_dtypes(np.number)
X_test_num  = X_test_df.select_dtypes(np.number)
X_train_cat = X_train_df.select_dtypes(np.object)
X_test_cat  = X_test_df.select_dtypes(np.object)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
transformer = MinMaxScaler().fit(X_train_num) 
X_train_scaled_arr = transformer.transform(X_train_num)
X_train_scaled = pd.DataFrame(X_train_scaled_arr, columns=X_train_num.columns)
```

```python
X_test_scaled_arr = transformer.transform(X_test_num)
X_test_scaled = pd.DataFrame(X_test_scaled_arr, columns=X_test_num.columns)
```

```python
encoder = OneHotEncoder(drop='first').fit(X_train_cat)
encoded_cat = encoder.transform(X_train_cat).toarray()
cols = encoder.get_feature_names(input_features=X_train_cat.columns)
onehot_encoded_train = pd.DataFrame(encoded_cat, columns=cols)
```

```python
encoded_test_cat = encoder.transform(X_test_cat).toarray()
onehot_encoded_test = pd.DataFrame(encoded_test_cat, columns=cols)
```

```python
X_train_treated = pd.concat([X_train_scaled, onehot_encoded_train], axis=1)
X_test_treated = pd.concat([X_test_scaled, onehot_encoded_test], axis=1)
```

```python
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True) 
```

```python
train = pd.concat([X_train_treated, y_train],axis=1)
```

```python
from sklearn.utils import resample
```

```python
no_donor = train[train['TARGET_B']==0]
yes_donor = train[train['TARGET_B']==1]

yes_donor_oversampled = resample(yes_donor, 
                                    replace=True, 
                                    n_samples = len(no_donor))

train_oversampled = pd.concat([no_donor,yes_donor_oversampled],axis=0)

y_train_over = train_oversampled['TARGET_B'].copy()
X_train_over = train_oversampled.drop('TARGET_B',axis = 1).copy()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20, 
                             max_samples=0.8,
                            n_jobs = -1)
                        
clf.fit(X_train_over, y_train_over)
print(clf.score(X_train_over, y_train_over))
print(clf.score(X_test_treated, y_test))


```

```python
cross_val_scores = cross_val_score(clf, X_train_over, y_train_over, cv=10, n_jobs = -1)
print(np.mean(cross_val_scores))
```
